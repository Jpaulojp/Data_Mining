```python
import os 
import numpy as np
import pandas as pd
```

```python
#Load dataset
airlines = pd.read_csv('Airlines.csv')
airlines.head().T
```

```python
airlines.dtypes
```

1 -  Qual a empresa aérea que possui maior quantidade de voos?

```python
gpb_airlines = airlines.groupby(['Airline'])['Flight'].sum().reset_index()
#print(gpb_airlines)
ordened = gpb_airlines.sort_values('Flight', ascending = False)
print('A empresa OO possui a maior quantidade de voos')
print(ordened)
```

2 - Quais as malhas (Airport From e Airport To) que levam o tempo entre 500 e 1000?

```python
fromto= airlines[(airlines['Time']>499) & (airlines['Time']<1001)]
r_time = fromto.groupby(['AirportFrom', 'AirportTo'])['Time'].sum().reset_index()
r_time.sort_values('Time', ascending = False)
```

```python
print('Numero de malhas total que levam o tempo de 500 a 1000:',len(fromto.index))
print('Numero de malhas sem repetição que levam o tempo de 500 a 1000:',len(r_time.index))
```

3 -  Quais as malhas (Airport From e Airport To)  que possuem a maior quantidade de voos? Apresentar ordenado do maior para o menor

```python
teste = airlines.sort_values('Flight', ascending = False)
teste2 = airlines.groupby(['AirportFrom', 'AirportTo'])['Flight'].sum().reset_index()
teste2.sort_values('Flight', ascending = False)
```

4 - Qual a empresa aérea que possui a maior quantidade de voos com atraso?

```python
delay = airlines.groupby(['Airline'])['Delay'].sum().reset_index()
print('A empresa que possui maior quantidade de atrasos é WN')
delay.sort_values('Delay', ascending = False)
```
